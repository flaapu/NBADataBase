```python
#Ejercicio 1 
#Cree una función en Python para calcular la cantidad de números pares en una lista determinada
def contador_par(lista):
    par = 0
    for n in lista:
        if n%2 == 0:
            par += 1
    return par

list1 = [2, 2, 2, 2, 2, 2, 5, 5, 1, 3, 5, 67, 82, 4, 5, 7, 1000000, 22, 3]
list3 = [2, -11, 5, 5, 1, -11, -1, 67, 82, 4, 5, 7, 0.3, 22, 3]

resultado = contador_par(list3)
resultado = contador_par(list1)

print("La cantidad de pares es: ", resultado)
```

```python
#Ejercicio 2
#Desarrolle una función en Python para calcular el desvío estándar de un arreglo numérico genérico
def DesvioEstandar(entrada):
  sumando = 0
  media = 0

  for n in entrada:
    sumando += n

  tamano = len(entrada) 
  media = sumando / tamano
  sumatoria = 0

  for n in entrada:
    sumatoria += pow((n - media),2)

  s2 = sumatoria / (tamano - 1)
  s = pow(s2,0.5)
  
  return s

entrada1 = [2, 3, 4, 5, 6, 1, 1, 1, 1, 2, 3, 4, 6, 9]
entrada2 = [2.3, 2.3, 2.3]
entrada3 = [2.3, 4, 4, 4, 4, 4, 4, 4, 4, 99, 99, 1, -5, -99]

print(DesvioEstandar(entrada1))
print(DesvioEstandar(entrada2))
print(DesvioEstandar(entrada3))




```

```python
# Realizar un script en Python usando funciones que generen tres listas de números 
# aleatorios de tamaños 10, 1000, y 10000, comprendidos entre [0,1], [5,8], [9,99]
# respectivamente

import random

def GenerarLista(cotaInf, cotaSup, long):
  lista = []
  for n in range(0, long):
    lista.append(random.randint(cotaInf, cotaSup))
  return lista


lista1 = GenerarLista(0, 1, 10)
lista2 = GenerarLista(5, 8, 1000)
lista3 = GenerarLista(9, 99, 10000)

print("Listas randoms 1", lista1)
print("Listas randoms 2", lista2)
print("Listas randoms 3", lista3)
```
